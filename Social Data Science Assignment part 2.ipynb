## Merge the data

- Merge two data frames
- Answer the following questions
  - Calculate the deaths/cases per 1000 population
  - Find the maximum of highest deaths/cases per 1000 population and their county names 
- Aggregate the data at the state level and find the following
  - Calculate the number of total cases/deaths per 1000 population
  - Which state had the highest cases/deaths, how many cases/deaths?


```python
#Merging the Two Dataframes 

covid_file.rename(columns = {'fips':'CountyId'}, inplace = True) 

df_merge = pd.merge(covid_file, county_file, on = "CountyId")

df_merge.drop("State", inplace = True, axis = 1)
df_merge.drop("County", inplace = True, axis = 1)

df_merge
```

```python
#adding deaths per 1000 of the population per county column

df_merge["deaths per 1000"] =  df_merge["deaths"] / (df_merge["TotalPop"] / 1000)  

df_merge
```

```python
#sorting data to find the county with the highest deaths per 1000

deaths_sorted = df_merge.sort_values("deaths per 1000", ascending = False)
deaths_sorted

#Hancock county in Georgia has the highest covid deaths per 1000 at 4.961
```

```python
#number of cases per 1000 in each county

df_merge["cases per 1000"] =  df_merge["cases"] / (df_merge["TotalPop"] / 1000)  

df_merge
```

```python
#sorting data to find the county with the highest cases per 1000

cases_sorted = df_merge.sort_values("cases per 1000", ascending = False)
cases_sorted

#Trousdale county in Tennessee has the highest covid case rate per 1000 at 191.041
```

```python
#aggregating deaths per 1000 to the state level 

grouped_data_deathsper1000_state = df_merge[['state', 'deaths per 1000']].groupby('state').mean()
max_deathsper1000 = grouped_data_deathsper1000_state.sort_values("deaths per 1000", ascending = False)
max_deathsper1000

#New Jersey has the highest deaths per 1000 at 1.589
```

```python
#aggregating cases per 1000 to the state level 

grouped_data_casesper1000_state = df_merge[['state', 'cases per 1000']].groupby('state').mean()
max_casesper1000 = grouped_data_casesper1000_state.sort_values("cases per 1000", ascending = False)
max_casesper1000

#Louisiana has the highest deaths per 1000 at 38.253
```
